```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.fftpack import fft
from numpy.fft import fft,ifft,fftshift,fftfreq
%matplotlib inline
```

```python
def mysinplot(fsamp,fnet,num):
    fsamp = float(fsamp)
    fnet = float(fnet)
    num = float(num)
    tot_time = num*(1/fnet)
    tot_samp = tot_time*fsamp
    arr = np.arange(0,tot_time,1/fsamp)
    y_axis = np.sin(2*np.pi*fnet*arr)
    plt.plot(arr,y_axis)
```

```python
def myctft(f,fs,T,T1):
#   sin wave will be there for minimum range from -T to T or -T1 to T1
    right_side = min(T,T1);
    left_side = -1*right_side;
    
#   'time' is to get sine wave whereas 'time2' is to plot signal. 'time' and 'time2' will be different in case t1 > t
    time = np.arange(left_side,right_side + 1/fs,1/fs);
    time2 = np.arange(-T1,T1+1/fs,1/fs);
    signal = np.sin(time*f*2*np.pi);
    
#   if T1 > T then append and prepend zeros to signal
    if(T1 > T):
        while time2.size != signal.size:
            signal = np.concatenate([[0],signal,[0]]);
            
#   Now do fft of signal and adjust its amplitude
    fft_sig = fft(signal);
    fft_sig = 2*fft_sig/signal.size;
    fft_sig = np.fft.fftshift(fft_sig);
    
#   Now x-axis of amplitude spectrum will be [-fs/2,fs/2]
    N = len(fft_sig);
    freqHZ = np.arange(-1*signal.size/2,signal.size/2);
    freqHZ = (freqHZ*fs)/signal.size;
    
#   plot signal
    plt.figure()
    plt.plot(time2,signal)
    plt.legend("Signal")
    plt.xlabel("Time in seconds")
    plt.ylabel("Amplitude")
    
#   plot fft
    plt.figure()
    plt.plot(freqHZ,abs(fft_sig))
    plt.legend("Fourier Transform of Signal")
    plt.xlabel("Frequency in Hertz")
    plt.ylabel("Amplitude")
```
