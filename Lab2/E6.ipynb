```python

%reset
from matplotlib import pyplot as plt
import numpy as np
import math
import random  
from datetime import datetime
import statistics as st
import pandas as pd

random.seed(datetime.now())
N = 10000;

a = [[0 for x in range(0)] for y in range(10)]
b = [];
means = [];
to_plot = [];
cov = [[0 for x in range(10)] for y in range(10)]

for i in range(0,N):
    tetha = random.uniform(-np.pi,np.pi);
    time = np.arange(0,10,1);
    vals = np.cos(0.2*np.pi*time + tetha);
    b.append(vals);

for i in range(0,N):
    to_plot.append(b[i][0]);

for n in range(0,10):
    val = 0;
    for angle in range(0,N):
        val += b[angle][n];
    val = val/N;
    means.append(val);

for n1 in range(0,10):
    for n2 in range(0,10):
        val = 0;
        for angle in range(0,N):
            val += (b[angle][n1] * b[angle][n2]);
        val = val/N;
        a[n1-n2].append(val);
        cov[n1][n2] = val;

print("Here is Rx function table :");
cov_1=pd.DataFrame(cov)
print(cov_1)

#print(cov);

print("\nHere is modified Rx function table :");
a_1=pd.DataFrame(a)
print(a_1)

#for i in range(0,10):
#    print(a[i]);
    
print("\nHere are means :");
means_1=pd.DataFrame(means)
print(means_1)
#for i in means:
#    print(i);
    
plt.hist(to_plot,bins = [-1,-0.9,-0.8,-0.7,-0.6,-0.5,-0.4,-0.3,-0.2,-0.1,0,0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90,1.00]);
plt.title("histogram");
plt.show();
```
